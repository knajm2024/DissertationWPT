```python
import pandas as pd
import matplotlib.pyplot as plt

excel_file = pd.ExcelFile('Paignton Zoo Footfall 2018 2024.xlsx')


summary_df = pd.read_excel(excel_file, sheet_name='Paignton Summary')


summary_df = summary_df.iloc[4:, [0, 1, 14, 15, 16]]  # Select relevant rows and columns
summary_df.columns = ['Month', 'Admissions Adult', 'Budget', 'Variance to Budget', 'Variance to Budget %']


summary_df.reset_index(drop=True, inplace=True)


summary_df.dropna(how='all', inplace=True)


summary_df['Admissions Adult'] = pd.to_numeric(summary_df['Admissions Adult'], errors='coerce')
summary_df['Budget'] = pd.to_numeric(summary_df['Budget'], errors='coerce')
summary_df['Variance to Budget'] = pd.to_numeric(summary_df['Variance to Budget'], errors='coerce')
summary_df['Variance to Budget %'] = pd.to_numeric(summary_df['Variance to Budget %'], errors='coerce')


summary_df['Month'] = pd.to_datetime(summary_df['Month'], errors='coerce', format='%b')


summary_df.dropna(inplace=True)


summary_df = summary_df.groupby(summary_df['Month'].dt.strftime('%b')).sum().reset_index()


print(summary_df.head())


plt.figure(figsize=(14, 7))
plt.plot(summary_df['Month'], summary_df['Admissions Adult'], label='Actual Visitors', marker='o')
plt.plot(summary_df['Month'], summary_df['Budget'], label='Budgeted Visitors', marker='o')
plt.fill_between(summary_df['Month'], summary_df['Admissions Adult'], summary_df['Budget'], color='gray', alpha=0.2)
plt.xlabel('Month')
plt.ylabel('Visitor Numbers')
plt.title('Actual vs Budgeted Visitor Numbers')
plt.legend()
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(14, 7))
plt.bar(summary_df['Month'], summary_df['Variance to Budget'], color='orange')
plt.xlabel('Month')
plt.ylabel('Variance to Budget')
plt.title('Variance to Budget')
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(14, 7))
plt.bar(summary_df['Month'], summary_df['Variance to Budget %'] * 100, color='green')
plt.xlabel('Month')
plt.ylabel('Variance to Budget (%)')
plt.title('Variance Percentage to Budget')
plt.xticks(rotation=45)
plt.show()
```
