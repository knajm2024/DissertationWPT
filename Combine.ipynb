```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


nz_file = 'Newquay Zoo Footfall 2018 2024.xlsx'
pz_file = 'Paignton Zoo Footfall 2018 2024.xlsx'
gdp_file = 'GDP98-21.xlsx'


nz = pd.read_excel(nz_file, sheet_name='Newquay Footfall')
pz = pd.read_excel(pz_file, sheet_name='Paignton Footfall')
gdp = pd.read_excel(gdp_file, sheet_name='in')


print("Newquay Zoo Footfall columns:", nz.columns)
print("Paignton Zoo Footfall columns:", pz.columns)
print("GDP Data columns:", gdp.columns)
print(gdp.head())

gdp['Year'] = pd.to_datetime(gdp['date']).dt.year
print(gdp.head())

#  Newquay Zoo
plt.figure(figsize=(10, 6))
sns.lineplot(data=nz, x='Date', y='Total', label='Newquay Zoo')
plt.title('Total Visitors Over Time - Newquay Zoo')
plt.xlabel('Date')
plt.ylabel('Total Visitors')
plt.legend()
plt.show()

#Paignton Zoo
plt.figure(figsize=(10, 6))
sns.lineplot(data=pz, x='Date', y='Total', label='Paignton Zoo')
plt.title('Total Visitors Over Time - Paignton Zoo')
plt.xlabel('Date')
plt.ylabel('Total Visitors')
plt.legend()
plt.show()


plt.figure(figsize=(10, 6))
sns.lineplot(data=gdp, x='Year', y='England', label='GDP per Head - England')  # Adjust 'England' to correct column
plt.title('GDP per Head Over Time - Cornwall and Devon')
plt.xlabel('Year')
plt.ylabel('GDP per Head')
plt.legend()
plt.show()
```
