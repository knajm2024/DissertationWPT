```python
import pandas as pd
data = pd.read_excel('Book1.xlsx')
```

```python
data.head(), data.info()
```

```python
import matplotlib.pyplot as plt


data.set_index('Area name', inplace=True)
ax = data.transpose().plot(figsize=(14, 8), legend=False)  # Turn off legend temporarily for positioning
plt.title('Trend of Unemployment Rates by Area')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.grid(True)  # Add grid for better readability


plt.legend(title='Area', loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust the plot area to make room for the legend
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

selected_areas = data.loc[data.index.str.contains("Cornwall|Devon")]

ax = selected_areas.transpose().plot(figsize=(14, 8), legend=False) 
plt.title('Trend of Unemployment Rates for Cornwall and Devon')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.grid(True)  

# Place a legend to the right of this plot
plt.legend(title='Area', loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.tight_layout(rect=[0, 0, 0.85, 1])  
plt.show()
```

```python
latest_year = data.columns[-1]
comparison = data[[latest_year]].sort_values(by=latest_year)
print(comparison)
```

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.heatmap(data.transpose(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Employment Rates')
plt.xlabel('Area')
plt.ylabel('Year')
plt.show()
```
