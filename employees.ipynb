```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "Year": [2018, 2019, 2020, 2021, 2022],
    "Average Headcount": [347, 358, 312, 267, 296],
    "Full-Time Equivalent": [266, 269, 233, 230, 246]
}


df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df["Average Headcount"], marker='o', label='Average Headcount')
plt.plot(df["Year"], df["Full-Time Equivalent"], marker='s', label='Full-Time Equivalent')
plt.xlabel('Year')
plt.ylabel('Number of Employees')
plt.title('WPT Employment Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting t
plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df["Average Headcount"], marker='o', label='Average Headcount')
plt.plot(df["Year"], df["Full-Time Equivalent"], marker='s', label='Full-Time Equivalent')
plt.xlabel('Year')
plt.ylabel('Number of Employees')
plt.title('WPT Employment Over Time')
plt.xticks(df["Year"].astype(int))  # Setting x-axis labels as integers
plt.legend()
plt.grid(True)


for i in range(len(df)):
    plt.text(df["Year"][i], df["Average Headcount"][i], df["Average Headcount"][i], ha='right')
    plt.text(df["Year"][i], df["Full-Time Equivalent"][i], df["Full-Time Equivalent"][i], ha='right')

plt.show()
```
